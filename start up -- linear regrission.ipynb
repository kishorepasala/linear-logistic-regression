```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics import r2_score
```

```python
start_df = pd.read_csv('50_Startups.csv')
```

```python
le = LabelEncoder()
start_df['State'] = le.fit_transform(start_df['State'])
```

```python
min_max = MinMaxScaler()
start_df['R&D Spend'] = min_max.fit_transform(np.array(start_df['R&D Spend']).reshape(-1,1))
start_df['Administration'] = min_max.fit_transform(np.array(start_df['Administration']).reshape(-1,1))
start_df['Marketing Spend'] = min_max.fit_transform(np.array(start_df['Marketing Spend']).reshape(-1,1))
```

```python
y = start_df['Profit']
start_df.drop(columns = ['Profit'], inplace = True)
```

```python
start_df.drop(columns = ['state'], inplace = True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(start_df, y, test_size = 0.25, random_state = 2)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
reg.score(x_test,y_test)
```

```python
print(
  'mean_squared_error : ', mean_squared_error(y_test, y_pred))
print(
  'mean_absolute_error : ', mean_absolute_error(y_test, y_pred))
```

```python
r2 = r2_score(y_test, y_pred)
```
